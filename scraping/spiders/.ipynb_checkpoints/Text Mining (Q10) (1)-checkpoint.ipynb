```python
# Importing necessary library
import pandas as pd
import numpy as np
import nltk
import os
import nltk.corpus

import seaborn as sb
import matplotlib.pyplot as plt
```

```python
rev = pd.read_csv('review_resolve_emoji.csv').drop(['Unnamed: 0'], axis=1)
rev.head()
```

```python
#extracting relevant columns 

revrev = rev.filter(['What is your name?','What can we do to make the event more engaging, appealing, or exciting?'], axis=1)
#renaming columns
SIC.columns = ["Name","Text"]
```

```python
SIC.head()
```

```python
#getting a list with all the texts forming separate sentences
text_list = SIC['Text'].tolist()
text_list
```

```python
import re
#remove punctuation with regex
all_words=[re.sub(r"[^a-zA-Z0-9]+", ' ', k) for k in text_list]
```

```python
all_words
```

```python
# importing word_tokenize from nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
#create a list to store the tokens
all_tokens=[]

# Passing the list elements word tokenize for breaking the sentences 
for text in all_words:
    all_tokens.append(word_tokenize(text))


all_tokens
```

```python
#flatten the list (turn list of lists into simple list)
all_tokens = [y for x in all_tokens for y in x]
all_tokens
```

```python
#convert to lowercase
all_tokens=[x.lower() for x in all_tokens]
all_tokens
```

```python
#following cells are manual data cleaning by observing what words are just for emphasis
#remove 'more' bcs it is implied in the sentences 
try:
    while True:
        all_tokens.remove('more')
except ValueError:
    pass

#remove perhaps
try:
    while True:
        all_tokens.remove('perhaps')
except ValueError:
    pass

#remove maybe
try:
    while True:
        all_tokens.remove('maybe')
except ValueError:
    pass

#remove doesn't
try:
    while True:
        all_tokens.remove('doesn\'t')
except ValueError:
    pass
```

```python
#replace "person" with "offline" (in-person is synonym of offline)
all_tokens=['offline' if x=='person' else x for x in all_tokens]
```

```python
all_tokens
```

```python
nltk.download('stopwords')
```

```python
#removing stop words i.e. words like "the", "or", "and" etc
from nltk.corpus import stopwords
a = set(stopwords.words('english'))
from itertools import cycle

stopwords=[]
stopwords = [x for x in all_tokens if x not in a]
# print(stopwords)
stopwords
```

```python
#generate word cloud
import matplotlib.pyplot as plt
import wordcloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

#convert list to string and generate
unique_string=(" ").join(stopwords)
#print(unique_string)
wordcloud = WordCloud(width = 1000, height = 500, max_words=50,background_color="white").generate(unique_string)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("wordcloud_q10"+".png", bbox_inches='tight')
plt.show()
plt.close()
```
