```python
mystr = 'I like this girl called kim-Go-Eun'
# doc = nlp('and kim go Eun as Eun-Tak are super cute')

casts = ['Gong Yoo', 'Kim Shin', 'Kim Go Eun', 'Ji Eun Tak', 'Lee Dong Wook', 'Yoo In Na', 'Yook Sung Jae', 'Yoo Duk Hwa']
# 'Grim Reaper', 'Sunny'
import spacy
from spacy.matcher import Matcher
from spacy.lang.en import English
nlp = spacy.load('en_core_web_lg')
@English.component("replace_cast")
def replace_cast(doc):
    name_matcher = Matcher(nlp.vocab)
    for cast in casts:
        cast_pattern = []
        name = cast.split()
        if len(name)==2:
            char1 = name[0].lower()
            char2 = name[1].lower()
            pt1 = [{'LOWER': char1+char2}]
#             pt2 = [{'LOWER': char1[0]+char2[0]}]
            pt3 = [{'LOWER': char1}, {'TEXT': '-', 'OP': '?'}, {'LOWER': char2}] 
            cast_pattern.append(pt1)
            cast_pattern.append(pt2)
            cast_pattern.append(pt3)
            name_matcher.add(cast, cast_pattern)
        elif len(name)==3:
            char1 = name[0].lower()
            char2 = name[1].lower()
            char3 = name[2].lower()
            pt1 = [{'LOWER': char1, 'OP': '?'}, {'LOWER': char2+char3}] 
            pt2 = [{'TEXT': char1[0].upper()+char2[0].upper()+char3[0].upper()}]
            pt3 = [{'LOWER': char1, 'OP': '?'}, {'TEXT': '-', 'OP': '?'}, {'LOWER': char2}, {'TEXT': '-', 'OP': '?'}, {'LOWER': char3}]
            cast_pattern.append(pt1)
            cast_pattern.append(pt2)
            cast_pattern.append(pt3)
            name_matcher.add(cast, cast_pattern)
    count = 1
    name_matches = name_matcher(doc)
    while count <= len(name_matches):
        match_id, start, end = name_matches[count-1]
        rep = nlp.vocab.strings[match_id]
        newDoc = doc[:start].text+' '+rep+' '+doc[end:].text 
        doc = nlp.make_doc(newDoc.strip()) # f" {replacement}"
        name_matches = name_matcher(doc)
        if len(rep.split())==2: # if name is two chars, +1
            count += 1
        else:
            count += 2
    return doc
nlp.add_pipe('replace_cast', first=True)
# 我们先replace第一个match，再match的时候前两个一定是刚刚修改过的match。第三个
# 是新的match。再修改它，再match的时候前四个是刚才修改的match。看第五个，没有第五个match了，不需要再修改了
```
